```python
from pathlib import Path
import pandas as pd
from IPython.display import display, HTML
from collections import Counter
import matplotlib.pyplot as plt
```

```python
def display_metadata(metadata):
    metadata = metadata.sort_values(by=["ITEM_ID"])
    num_rows = len(metadata)

    df_describe = metadata.describe()
    df_nunique = metadata.nunique()

    display(HTML(f"<h3>Metadata</h3><h4>Number of rows: {num_rows}</h4>"))
    display(HTML("<h4>Metadata head</h4>"))
    display(metadata)
    display(HTML("<h4>Metadata describe</h4>"))
    display(df_describe.T)
    display(HTML("<h4>Metadata nunique</h4>"))
    display(pd.DataFrame({"count": df_nunique}).T)
```

```python
envato_file_metadata = "../envato_fonts/metadata/fonts_file_level_metadata.csv"
file_meatadata = pd.read_csv(envato_file_metadata)
print(f"metadata: {envato_file_metadata}")
print(f"file_meatadata.shape: {file_meatadata.shape}")
display_metadata(file_meatadata)

envato_zip_metadata = "../envato_fonts/metadata/fonts_zip_level_metadata.csv"
zip_metadata = pd.read_csv(envato_zip_metadata)
print(f"metadata: {envato_zip_metadata}")
print(f"zip_metadata.shape: {zip_metadata.shape}")
display_metadata(zip_metadata)
```

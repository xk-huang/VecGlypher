```plaintext
from pathlib import Path
import numpy as np
import cv2
import matplotlib.pyplot as plt


file="../baseline_data/deepvecfont_v2-data/deepvecfont_v2/data/vecfont_dataset/eng/test/0000/rendered_64.npy"

data = np.load(file)

concate_img = data[0]
for i in range(1, len(data)):
    concate_img = np.concatenate((concate_img, data[i]), axis=1)

save_img_path = "misc/deepvecfont_v2-data.png"
cv2.imwrite(save_img_path, concate_img)
print("save to", save_img_path)
```

```plaintext
base_dir="data/baseline_data/deepvecfont_v2-data/deepvecfont_v2/data/vecfont_dataset/eng/test/0000/"
base_dir = Path(base_dir)
files = [
    "class.npy", "font_id.npy", "pts_aux.npy", "rendered_64.npy", "seq_len.npy", "sequence_relaxed.npy", "sequence.npy"
]
data_list = []
for file in files:
    file_path =base_dir / file
    data = np.load(file_path)
    data_list.append(data)

for file, data in zip(files, data_list):
    print("\n")
    print(file, data.shape, len(data.shape))
    if len(data.shape) == 0:
        print(file, data)
    elif len(data.shape) == 2:
        print(file, data[0, :3])
    elif len(data.shape) == 3:
        print(file, data[0, :3, :3])
    else:
        raise ValueError(f"data shape: {data.shape}")
```

```python
import glob
import json
from pathlib import Path

import matplotlib.pyplot as plt

import pandas as pd
import seaborn as sns
```

```python
content_file = "data/processed/content/alphanumeric.txt"
content_file = Path(content_file)

data = []
with open(content_file, "r") as f:
    for line in f:
        line = line.strip()
        if line:
            data.append(line)

len_list = [len(x)  for x in data]

fig, ax = plt.subplots(figsize=(6, 4))
ax.hist(len_list, bins=100)
ax.set_title(f"Content length distribution: {content_file.stem}", fontdict={"fontsize": 10})
ax.set_xlabel("Content length")
ax.set_ylabel("Frequency")
plt.show()
```
